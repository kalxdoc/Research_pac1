```python
from numpy import mean
import pandas as pd
from sklearn.impute import SimpleImputer
df=pd.read_csv("/content/dhaka_soil_data.csv")
print("--1.First row")
display(df.head())
print("--- 1. IDENTIFY MISSING VALUES ---")
# .isnull().sum() is your best friend. It counts the exact number of
# missing cells in every single column.
missing_counts = df.isnull().sum()
print(missing_counts)
print("-" * 30)
print("\n---SKlearm method---")
df_sklearn=df.copy()
imputer=SimpleImputer(strategy='mean')
df_sklearn[['Distance_to_Tannery_m']]=imputer.fit_transform(df_sklearn[['distance']])
```

```python
from numpy import mean
import pandas as pd
from sklearn.impute import SimpleImputer
df=pd.read_csv("/content/dhaka_soil_data.csv")
print("--1.First row")
display(df.head())
print("--- 1. IDENTIFY MISSING VALUES ---")
# .isnull().sum() is your best friend. It counts the exact number of
# missing cells in every single column.
missing_counts = df.isnull().sum()
print(missing_counts)
print("-" * 30)
print("\n---SKlearm method---")
df_sklearn=df.copy()
imputer=SimpleImputer(strategy='mean')
df_sklearn[['Distance_to_Tannery_m']]=imputer.fit_transform(df_sklearn[['Distance_to_Tannery_m']])
print(f"missin values in column: {df_sklearn['Distance_to_Tannery_m'].isnull().sum()}")
print(f"Calculated Mean Distance: {imputer.statistics_[0]:.2f} meters")
```
