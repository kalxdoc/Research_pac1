```python
#Convert any categorical text data (like "Industrial Area" or "Residential Area") into numbers using LabelEncoder or OneHotEncoder
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
df=pd.read_csv("/content/dhaka_soil_data.csv")
print("--- ORIGINAL DATA ---")
display(df[['Sample_ID', 'Land_Use']].head())
print("\n--- METHOD A: Label Encoding (For Ranked Data) ---")
# 1. Initialize the encoder
label_enc = LabelEncoder()

# 2. Fit and transform the column
df_label = df.copy()
df_label['Land_Use_Encoded'] = label_enc.fit_transform(df_label['Land_Use'])
display(df_label[['Sample_ID', 'Land_Use', 'Land_Use_Encoded']].head())
print("\n--- METHOD B: One-Hot Encoding via Pandas (For Unranked Data) ---")
# Pandas has a built-in function called 'get_dummies' which is the easiest
# way for beginners to do One-Hot Encoding.
df_onehot = pd.get_dummies(df, columns=['Land_Use'], dtype=int)

# Notice how the original 'Land_Use' column is gone, replaced by new binary columns!
display(df_onehot.head())
```
